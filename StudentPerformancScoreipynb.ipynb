```python
from torch.utils.data import DataLoader, Dataset
```

```python
class Student_Dataset(Dataset):
  def __init__(self,mode='train'):
    super(Student_Dataset,self).__init__()
    #data preprocessing task
    student_data = pd.read_csv('student.csv')
    student_data.dropna()
    student_data = student_data.drop_duplicates()
    student_data = student_data.rename(columns = {'Hours Studied':'Hours',
                          'Previous Scores': 'PREV',
                          'Extracurricular Activities':'ACV',
                          'Sleep Hours': 'HRS',
                          'Sample Question Papers Practiced':'QS',
                          'Performance Index':'Score'})
    label_encoder = preprocessing.LabelEncoder()
    student_data['ACV'] = label_encoder.fit_transform(student_data['ACV'])
    scaler_x = preprocessing.MinMaxScaler()
    scaler_y = preprocessing.MinMaxScaler()
    X = student_data.iloc[:,0:-1].values
    Y = student_data.iloc[:,-1].values
    Y = Y.reshape(-1,1)


    X_train,X_test,Y_train,Y_test = train_test_split(X,Y)

    self.X_train = scaler_x.fit_transform(X_train)
    self.X_test = scaler_x.transform(X_test)
    self.Y_train = scaler_y.fit_transform(Y_train)
    self.Y_test = scaler_y.fit_transform(Y_test)

    if mode == "train":
      self.features = self.X_train
      self.target = self.Y_train
    else:
      self.features = self.X_test
      self.target = self.Y_test

  def __len__(self):
    return len(self.features)

  def __getitem__(self,pos):
    return self.features[pos],self.target[pos]

```

```python
train_dataset = Student_Dataset(mode='train')
```

```python
train_dataset.__getitem__(4)
```

```python
train_loader = DataLoader(train_dataset, batch_size = 32)
```

```python
fg = iter(train_loader)
```

```python
next(fg)[1].shape
```

```python
import torch.nn as nn
```

```python
## ANN Used
##Input Layer, Hidden Layer, Output Layer
## Module indicates this is a neural network class
class STUDENT_PREDICTION(nn.Module):
  def __init__(self, input_features = 5):
    super(STUDENT_PREDICTION,self).__init__()

   ## Create a Linear Layer
   ## fc = fully connected linear layer
   ## Input linear layer
   ## Input neurons 5, output neuron/hidden layer is 4
    self.fc1 = nn.Linear(input_features,4)
    ## Hidden layer is 4 output layer is 1 / regression output and prediction
    self.fc2 = nn.Linear(4,4)
    self.fc3 = nn.Linear(4,1)

  def forward(self, X):
    ##X -> (32,5)
    X = self.fc1(X) # (32,5) -> (32,4)
    X = self.fc2(X) # (32,4) -> (32,4)
    X = self.fc3(X) #(32,4) -> (32,1)
    return X
```

```python
model = STUDENT_PREDICTION()
```

```python
import torch
```

```python
## If the loss is high the model is not doing good
## if its low thats good
loss_func = torch.nn.MSELoss()
##
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)
```

```python
#data for testing mode
test_dataset = Student_Dataset(mode = 'test')
test_loader = DataLoader(test_dataset,
                          batch_size = 32)
```

```python
loss_arr  = []
test_loss_arr = []
```

```python
EPOCHS = 200
for ep in range(EPOCHS):
  model.train()
  for i,(features,target) in enumerate(train_loader):
    # (32,5) target = (32,1)
    pred = model.forward(features.float())
    # pred -> (32,1)
    optimizer.zero_grad()
    loss = loss_func(target.float(),pred)
    loss.backward()
    optimizer.step()


  model.eval()
  #testing_iteration
  for i,(features,target) in enumerate(test_loader):
    # (32,5) target = (32,1)
    pred = model.forward(features.float())
    # pred -> (32,1)
    test_loss = loss_func(target.float(),pred)


  print(f'Epoch {ep} Train Loss = {loss.detach().cpu().numpy()}  Test Loss = {test_loss.detach().cpu().numpy()}')
  loss_arr.append(loss.detach().cpu().numpy())
  test_loss_arr.append(test_loss.detach().cpu().numpy())
```
